## Bucle while
[al indice](#Contenidos)  

Se trata de otra manera de implementar un bucle en programacin. Los bucles tienen que ir siempre limitados. En el caso del `for`, le ponamos un nmero concreto de ejecuciones, segn el iterable que estuvisemos recorriendo. Para el `while` es algo diferente. Tiene una **condicin de ejecucin**, que mientras que se cumpla (`True`), seguir ejecutando una y otra vez. Por otro lado, el bucle tiene una **variable de ejecucn**, al igual que en el `for`, que se ir actualizando con cada vuelta, y es esa variable la que determina cundo acaba el bucle.



**Cuidado** con estos bucles ya que es muy fcil olvidarnos de actualiza la variable de ejecucin, o equivocarnos en la condicin de ejecucin. Si esto ocurre el cdigo se quedar corriendo hasta que detengamos el kernel (botn *interrupt the kernel*, arriba al lado del Run)

Veamos un ejemplo.

````python
from time import time
````

La manera ms habitual de implementar estos bucles es:
1. Declaro la **variable de ejecucin fuera del bucle**
2. Establezco una **condicin de ejecucin** para determinar cundo queremos que se pare el bucle.
3. **Actualizo la variable de ejecucin** en cada iteracin del bucle.

## Break/continue
[al indice](#Contenidos)  

Son dos sentencias que podemo usar dentro de los bucles para evitar ejecutar cdigo de ms.

### Break
Se usa cuando queremos salir del bucle forzadamente. Imagina que eres una tienda y ests buscando con un bucle for si al menos uno de los pedidos era un abrigo. Si has tenido 1000 pedidos, vas a tener que iterar sobre todos y mediante un `if`, comprobar si es un abrigo. Ahora bien, si el abrigo es el primer elemento de la lista, el `for` va a recorrer igualmente los otros 999 elementos, cuando no es necesario. Con un `break` podremos salirnos del bucle y continuar con el programa.



### Continue
Esta sentencia se usa dentro de un bucle para indicarle que continue con el siguiente elemento del iterable. Al igual que con el `break`, nos sirve para evitar que se ejecute cdigo de ms. Volviendo al ejemplo anterior, si despus de comprobar que tenemos un abrigo, hay 200 lneas ms de cdigo que se utiliza en otros casos, con un `continue` evitamos que se ejecute todo eso, hacemos lo que tengamos que hacer con el abrigo, y le decimos al bucle que pase al siguiente elemento, e ignore el resto del cdigo.



**Los bucles `for` y `while`, as como `break` y `continue`, son sentencias complicadas de entender, y si es la primera vez que programas te va a suponer un cambio en la manera de pensar y de solucionar problemas, por ello te recomiendo que cojas papel y boli y hagas los primeros ejercicios de bucles viendo las iteraciones una a una y calculando manualmente todas las opeaciones de dentro del bucle.**

## Try/except
[al indice](#Contenidos)  

Qu ocurre cuando hay un error en nuestro cdigo? Se para toda la ejecucin. Por muy buenos programadores que seamos, hay que contar siempre con que puede haber errores. Si intuimos que el comportamiento de nuestro cdigo puede ser algo impredecible, en programacin podemos usar las sentencias `try/except` para capturar ese error, tomar decisiones, y que el cdigo pueda continuar ejecutndose.

La sintaxis es la siguiente:

> ```Python
> try:
>     Cdigo que puede contener errores
> except:
>     Qu hacer si nos encontramos con un error
> ```

Tambin puede ser de utilidad conocer el mensaje de error sin que el programa termine

## Resumen

````python
# If/elif/else
mi_nota_de_examen = 7

if mi_nota_de_examen < 5 :
    print("A septiembre :(")

elif mi_nota_de_examen < 6 :
    print("Suficiente")
    
elif mi_nota_de_examen < 7 :
    print("Bien")
    
elif mi_nota_de_examen < 9 :
    print("Notable")
    
else:
    print("Sobresaliente")
    
    
# Bucle for
dias_semana = ["Lunes", "Martes", "Mircoles", "Jueves", "Viernes", "Sbado", "Domingo"]

for dia in dias_semana:
    print(dia)
    
    
# Bucle while
i = 0

while(i < 5):
    print(i)
    i = i + 1
    
    
# Break y continue
for val in "string":
    if val == "i":
        break
    print(val)

print("Fin")


# Try/except
try:
  print(variable_)
except:
  print("El codigo tiene errores porque la variable 'variable_' no existe")

print("Continuo con el programa")
````
